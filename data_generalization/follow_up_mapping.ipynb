```python
import os
import pandas as pd
from tqdm import tqdm
```

```python
OUT_PATH = 'nodes'
SCHEMAS = 'node_template/submission_follow_up_template.tsv'
```

```python
#read the diagnosis node header
follow_up_pd = pd.read_csv(SCHEMAS, sep = '\t')
follow_up_pd.drop(index=follow_up_pd.index[0], 
        axis=0, 
        inplace=True)
follow_up_pd
```

```python
#Read case node data
case_node_pd = pd.read_csv('nodes/case_node.tsv', sep = '\t')
case_node_pd = case_node_pd.iloc[: , 1:]
case_node_pd
```

```python
visit_pd = pd.read_csv('csv_datasets/VISIT.csv')
visit_pd
```

```python
#calculate the redcap_event_name time
def get_redcap_event_name(value):
    if 'Week' in value:
        to_day = int(value.split(' ')[-1]) * 7
        #print('Week_to_day: ', to_day)
        return to_day
    if 'Day' in value:
        to_day = int(value.split(' ')[-1]) * 1
        #print('Day_to_day: ', to_day)
        return to_day
    if 'Screening' in value:
        to_day = -1
        #print('Screening_to_day: ', to_day)
        return to_day
    
    ##need to be clarify
    if 'Medical Record Follow-up Arm' in value:
        #record_to_day = int(value.split(' ')[-1]) * 9999
        to_day = -1
        #print('Medical_Record_to_day: ', -1)
        return to_day

def get_cps(value):
    if value == 999:
        return None
    elif value == None:
        return 0
    else:
        return value

def row_init(len):
    row = [None]*len
    return row

def set_column_value(df, col_name, row, value):
    index= df.columns.get_loc(col_name)
    row[index] = value
    return row
    
    
```

```python
for index,row in tqdm(visit_pd.iterrows(),total=visit_pd.shape[0],desc="process Follow_up node"):
    #general properies for both observation and clinical trial study
    follow_up_type = 'follow_up'
    project_id = 'ardac-p1031'
    
    #map case.submitter.id
    case_submitter_id = row.usubjid
    for index_case,row_case in case_node_pd.iterrows():
        if str(case_submitter_id) in str(row_case['*submitter_id']):
            case_submitter_id = str(row_case['*submitter_id'])
            break
        
    
    days_to_follow_up = get_redcap_event_name(row.redcap_event_name)
    submitter_id = str(case_submitter_id) + '_' + str(days_to_follow_up)
    bmi = row.bmi
    #if cps = blank, should I assign 0 ?
    child_pugh_score = get_cps(row.cps)
    lille_score = row.lille
    maddreys_score = row.mdf
    meld_score = row.meld
    
    
    #Assign the value in to the follow up node
    follow_up_row = row_init(len(follow_up_pd.axes[1]))
    follow_up_row = set_column_value(follow_up_pd, '*type', follow_up_row, follow_up_type)
    follow_up_row = set_column_value(follow_up_pd, 'project_id', follow_up_row, project_id)
    follow_up_row = set_column_value(follow_up_pd, '*submitter_id', follow_up_row, submitter_id)
    follow_up_row = set_column_value(follow_up_pd, 'cases.submitter_id', follow_up_row, case_submitter_id)
    follow_up_row = set_column_value(follow_up_pd, '*days_to_follow_up', follow_up_row, days_to_follow_up)
    follow_up_row = set_column_value(follow_up_pd, 'bmi', follow_up_row, bmi)
    follow_up_row = set_column_value(follow_up_pd, 'child_pugh_score', follow_up_row, child_pugh_score)
    follow_up_row = set_column_value(follow_up_pd, 'lille_score', follow_up_row, lille_score)
    follow_up_row = set_column_value(follow_up_pd, 'maddreys_score', follow_up_row, maddreys_score)
    follow_up_row = set_column_value(follow_up_pd, 'meld_score', follow_up_row, meld_score)
    follow_up_row = set_column_value(follow_up_pd, 'visit_day', follow_up_row, days_to_follow_up)
    
    
    follow_up_pd.loc[len(follow_up_pd.index)] = follow_up_row 
```
