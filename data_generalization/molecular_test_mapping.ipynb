```python
#Read follow up node data
follow_up_node_pd = pd.read_csv('nodes/follow_up_node.tsv', sep = '\t')
follow_up_node_pd
```

```python
#calculate the redcap_event_name time
def get_redcap_event_name(value):
    if 'Week' in value:
        week_to_day = int(value.split(' ')[-1]) * 7
        #print('Week_to_day: ', week_to_day)
        return week_to_day
    elif 'Day' in value:
        day_to_day = int(value.split(' ')[-1])
        #print('Day_to_day: ', day_to_day)
        return day_to_day
    elif 'Screening' in value:
        screening_to_day = 0
        #print('Screening_to_day: ', screening_to_day)
        return screening_to_day
    
    ##need to be clarify
    elif 'Medical Record Follow-up Arm' in value:
        #record_to_day = int(value.split(' ')[-1]) * 9999
        #print('Medical_Record_to_day: ', -1)
        return -1
    
def check_lab_result(labs_row, test_name):
    test_value = labs_row.iloc[0][test_name]
    return [test_name, test_value]
    
def row_init(len):
    row = [None]*len
    return row

def set_column_value(df, col_name, row, value):
    index= df.columns.get_loc(col_name)
    row[index] = value
    return row
    
```

```python
import math
labs_df["follow_up_submitter_id"] = np.nan
labs_df["days_to_follow_up"] = np.nan
for index,row in tqdm(labs_df.iterrows(),total=labs_df.shape[0],desc="process enrollment"):
    days_to_follow_up = str(get_redcap_event_name(row.redcap_event_name))
    case_submitter_id = row.usubjid
    follow_up_submitter_id = str(case_submitter_id) + '_' + days_to_follow_up
    labs_df.at[index, 'follow_up_submitter_id'] = follow_up_submitter_id
    labs_df.at[index, 'days_to_follow_up'] = days_to_follow_up
    
labs_df.drop(labs_df[labs_df['days_to_follow_up'] == -1].index, inplace = True)
labs_df
```

```python
import math 
test_count = 0
for index,row in tqdm(follow_up_node_pd.iterrows(),total=follow_up_node_pd.shape[0],desc="process Molecular_test node"):
    molecular_test_type = 'molecular_test'
    project_id = 'ARDaC-AlcHepNet' 
    molecular_test_follow_ups_submitter_id = row[2]
    #print(molecular_test_follow_ups_submitter_id)
    
    for index_lab,row_lab in labs_df.iterrows():
            patient_id = molecular_test_follow_ups_submitter_id.split('_')[0]
            followup_day = molecular_test_follow_ups_submitter_id.split('_')[-1]
            if str(row_lab.usubjid) == patient_id and str(row_lab.days_to_follow_up) == followup_day:
                labs_row = labs_df.iloc[[index_lab]]
                break
    
    #follow up does not have a molecular test result
    if labs_row.empty:
        test_count = 0
        continue
    else:
        test_list = ['inr','pt','albumin','bili_tot','bili_dir','creat','est_gfr','alt','ast','alkaline','protein_tot','bun']
        for item in test_list:
            if item != 'bun':
                lab_value = check_lab_result(labs_row, item)
                lab_result = check_lab_result(labs_row, item+'_norm')
            else:
                lab_value = check_lab_result(labs_row, item)
                lab_result = [np.nan,np.nan]
                
            if  lab_result[1] == lab_result[1] or lab_value[1] == lab_value[1]: #either test value or test result has a value, run code below.
                laboratory_test = lab_value[0]
                test_result = lab_result[1]
                test_value = lab_value[1]
                
                submitter_id = row[2] + '_' + str(laboratory_test)   #row[2] is '*submitter_id'
                #test_count += 1
                molecular_test_row = row_init(len(molecular_test_node.axes[1]))
                molecular_test_row = set_column_value(molecular_test_node, '*type', molecular_test_row, molecular_test_type)
                molecular_test_row = set_column_value(molecular_test_node, 'project_id', molecular_test_row, project_id)
                molecular_test_row = set_column_value(molecular_test_node, '*submitter_id', molecular_test_row, submitter_id)
                molecular_test_row = set_column_value(molecular_test_node, '*follow_ups.submitter_id', molecular_test_row, molecular_test_follow_ups_submitter_id)

                molecular_test_row = set_column_value(molecular_test_node, 'laboratory_test', molecular_test_row, laboratory_test)
                molecular_test_row = set_column_value(molecular_test_node, 'test_value', molecular_test_row, test_value)

                molecular_test_row = set_column_value(molecular_test_node, 'test_result', molecular_test_row, test_result)
                molecular_test_node.loc[len(molecular_test_node.index)] = molecular_test_row
        #test_count = 0  
```

```python
#add aki cyctatin-C into molecular_test
for index,row in tqdm(aki_cystatin_df.iterrows(),total=aki_cystatin_df.shape[0],desc="process aki barcode"):
    molecular_test_type = 'molecular_test'
    project_id = 'ARDaC-AlcHepNet' 
    
    pat_id = row['Subject ID']
    index_day = row['Last Time point for every subject']
    if 'Base line' in index_day:
        index_day = str(0)
    elif 'day' in index_day:
        index_day = str(index_day.split(' ')[0])
    molecular_test_follow_ups_submitter_id = str(pat_id) + '_clinical_' + index_day
    submitter_id = molecular_test_follow_ups_submitter_id + '_cystatin-c'
    laboratory_test = 'Cystatin-C'
    test_value = row['Plasma Conc (mg/L)']
    test_unit = 'mg/L'
    barcode = row['Barcode']
    
    molecular_test_row = row_init(len(molecular_test_node.axes[1]))
    molecular_test_row = set_column_value(molecular_test_node, '*type', molecular_test_row, molecular_test_type)
    molecular_test_row = set_column_value(molecular_test_node, 'project_id', molecular_test_row, project_id)
    molecular_test_row = set_column_value(molecular_test_node, '*submitter_id', molecular_test_row, submitter_id)
    molecular_test_row = set_column_value(molecular_test_node, '*follow_ups.submitter_id', molecular_test_row, molecular_test_follow_ups_submitter_id)
    molecular_test_row = set_column_value(molecular_test_node, 'laboratory_test', molecular_test_row, laboratory_test)
    molecular_test_row = set_column_value(molecular_test_node, 'test_value', molecular_test_row, test_value)
    molecular_test_row = set_column_value(molecular_test_node, 'test_unit', molecular_test_row, test_unit)
    molecular_test_row = set_column_value(molecular_test_node, 'aliquots.submitter_id', molecular_test_row, barcode)
    
    molecular_test_node.loc[len(molecular_test_node.index)] = molecular_test_row
```

```python
for index,row in tqdm(molecular_test_node.iterrows(),total=molecular_test_node.shape[0],desc="Cleaning Molecular Test Node"):
    if row.laboratory_test == 'albumin':
        molecular_test_node.at[index,'laboratory_test'] = 'Albumin'
    if row.laboratory_test == 'alkaline':
        molecular_test_node.at[index,'laboratory_test'] = 'Alkaline Phosphatase'
    if row.laboratory_test == 'alt':
        molecular_test_node.at[index,'laboratory_test'] = 'ALT'
    if row.laboratory_test == 'ast':
        molecular_test_node.at[index,'laboratory_test'] = 'AST'
    if row.laboratory_test == 'bili_dir':
        molecular_test_node.at[index,'laboratory_test'] = 'Direct Bilirubin'
    if row.laboratory_test == 'bili_tot':
        molecular_test_node.at[index,'laboratory_test'] = 'Total Bilirubin'
    if row.laboratory_test == 'bun':
        molecular_test_node.at[index,'laboratory_test'] = 'BUN'
    if row.laboratory_test == 'creat':
        molecular_test_node.at[index,'laboratory_test'] = 'Creatinine'
    if row.laboratory_test == 'est_gfr':
        molecular_test_node.at[index,'laboratory_test'] = 'Estimated GFR'
    if row.laboratory_test == 'inr':
        molecular_test_node.at[index,'laboratory_test'] = 'INR'
    if row.laboratory_test == 'protein_tot':
        molecular_test_node.at[index,'laboratory_test'] = 'Total Protein'
    if row.laboratory_test == 'pt':
        molecular_test_node.at[index,'laboratory_test'] = 'Prothrombin Time'
```
