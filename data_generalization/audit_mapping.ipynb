```python
#Read case node data
case_node_pd = pd.read_csv('/Users/nanxinjin/Documents/IU/gen3-aws/1031/case_node.tsv', sep = '\t')
case_node_pd
```

```python
visit_pd = pd.read_csv('csv_datasets/VISIT.csv')
visit_pd
```

```python
for index,row in tqdm(visit_pd.iterrows(),total=visit_pd.shape[0],desc="process visit table"):
        if row['redcap_event_name'] != 'Week 0' and row['redcap_event_name'] != 'Day 0':
            visit_pd.drop(index, inplace=True)
```

```python
visit_pd.reset_index(inplace = True, drop = True)
visit_pd
```

```python
def row_init(len):
    row = [None]*len
    return row

def set_column_value(df, col_name, row, value):
    index= df.columns.get_loc(col_name)
    row[index] = value
    return row
```

```python
for index,row in tqdm(case_node_pd.iterrows(),total=case_node_pd.shape[0],desc="process audit node"):
    usubjid = row['*submitter_id'].split('_')[0]
    
    for index_visit,row_visit in visit_pd.iterrows():
        if str(row_visit.usubjid) == usubjid:
            visit_row = visit_pd.iloc[[index_visit]]
            break
    audit_type = 'audit'
    project_id = 'ARDaC-AlcHepNet'
    cases_submitter_id = row['*submitter_id']
    submitter_id = cases_submitter_id+'_audit'
    auditnd = visit_row['auditnd'].item()
    adtcomploth = visit_row['adtcomploth'].item()
    adt0101 = visit_row['adt0101'].item()
    adt0102 = visit_row['adt0102'].item()
    adt0103 = visit_row['adt0103'].item()
    adt0104 = visit_row['adt0104'].item()
    adt0105 = visit_row['adt0105'].item()
    adt0106 = visit_row['adt0106'].item()
    adt0107 = visit_row['adt0107'].item()
    adt0108 = visit_row['adt0108'].item()
    adt0109 = visit_row['adt0109'].item()
    adt0110 = visit_row['adt0110'].item()
    
    audit_row = row_init(len(audit_pd.axes[1]))
    
    audit_row = set_column_value(audit_pd, '*type', audit_row, audit_type)
    audit_row = set_column_value(audit_pd, 'project_id', audit_row, project_id)
    audit_row = set_column_value(audit_pd, '*submitter_id', audit_row, submitter_id)
    audit_row = set_column_value(audit_pd, 'cases.submitter_id', audit_row, cases_submitter_id)
    audit_row = set_column_value(audit_pd, 'auditnd', audit_row, auditnd)
    audit_row = set_column_value(audit_pd, 'adtcomploth', audit_row, adtcomploth)
    audit_row = set_column_value(audit_pd, 'adt0101', audit_row, adt0101)
    audit_row = set_column_value(audit_pd, 'adt0102', audit_row, adt0102)
    audit_row = set_column_value(audit_pd, 'adt0103', audit_row, adt0103)
    audit_row = set_column_value(audit_pd, 'adt0104', audit_row, adt0104)
    audit_row = set_column_value(audit_pd, 'adt0105', audit_row, adt0105)
    audit_row = set_column_value(audit_pd, 'adt0106', audit_row, adt0106)
    audit_row = set_column_value(audit_pd, 'adt0107', audit_row, adt0107)
    audit_row = set_column_value(audit_pd, 'adt0108', audit_row, adt0108)
    audit_row = set_column_value(audit_pd, 'adt0109', audit_row, adt0109)
    audit_row = set_column_value(audit_pd, 'adt0110', audit_row, adt0110)
    
    
    audit_pd.loc[len(audit_pd.index)] = audit_row 
```

```python
#output the audit node
audit_node_path = os.path.join(OUT_PATH, 'audit.tsv')
audit_pd.to_csv(audit_node_path, sep = '\t', index=False)
audit_pd
```
